#### Pandas 데이터 합치기

(1) 옆으로 합치기 => merge   
(2) 위아래로 합치기 => concat   

데이터   
https://github.com/Play-with-data/datasalon

```python
import pandas as pd
import numpy as np
```

```python
# sample_1 데이터 불러오기
# header=1 : header를 0에서 1로 변경 / 2019년 11월 입국객 수는 컬럼 헤더가 아님
# skipfooter=2 : 받은 데이터 중에서 합계 등이 있으므로 총합계, 전년동기를 제거
sample_1 = pd.read_excel('https://raw.githubusercontent.com/Play-with-data/datasalon/master/02_%EA%B0%9C%EC%A0%95%ED%8C%90/2_Data_Analysis_Basic/files/sample_1.xlsx',
                         sheet_name='Sheet1', header=1, skipfooter=2)

sample_1['기준년월'] = '2019-11'
sample_1
```

```python
# 코드마스터 데이터 불러오기
code_master = pd.read_excel('https://raw.githubusercontent.com/Play-with-data/datasalon/master/02_%EA%B0%9C%EC%A0%95%ED%8C%90/2_Data_Analysis_Basic/files/sample_codemaster.xlsx')

code_master
```

```python
# merge : DB join과 동일

# how='left' : 왼쪽을 기준으로
# left_on='국적코드', right_on='국적코드' : 국적코드가 일치하는 것끼리

# left outer join
# merge_result = pd.merge(left=sample_1, right=code_master, how='left', left_on='국적코드', right_on='국적코드')

# inner join
merge_result = pd.merge(left=sample_1, right=code_master, how='inner', left_on='국적코드', right_on='국적코드')

merge_result
```

```python
# concat : DB Union
sample_2 = pd.read_excel('https://raw.githubusercontent.com/Play-with-data/datasalon/master/02_%EA%B0%9C%EC%A0%95%ED%8C%90/2_Data_Analysis_Basic/files/sample_2.xlsx',
                         header=1, skipfooter=2)
sample_2
```

```python
sample_2['기준년월'] = '2019-12'

sample_2_code = pd.merge(left=sample_2, right=code_master, how='left', left_on='국적코드', right_on='국적코드')

sample_2_code
```

```python
# append 함수가 없어짐 -> concat 사용
total_result = pd.concat([merge_result, sample_2_code])

total_result
```

```python
# 전치 행렬
total_result.T
```

```python
# 피벗 테이블 : 결과를 기준별로 다시 집계내는 함수
# index : 행
# columns : 열
total_result.pivot_table(values='입국객수', index='국적명', columns='기준년월')
```

```python
total_result.pivot_table(values='입국객수', index='성별', columns='국적명')
```
