##### Matplotlib 사용

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

````python
arr = np.random.randn(30)
arr
````

````python
# 단일 차트
plt.plot(arr)
````

````python
# 다중 차트
fig = plt.figure() # Figure 객체 생성
ax1 = fig.add_subplot(2, 2, 1) # 차트 1 (2행 2열 차트 중 1)
ax2 = fig.add_subplot(2, 2, 2) # 차트 2 (2행 2열 차트 중 2)
ax3 = fig.add_subplot(2, 2, 4) # 차트 3 (2행 2열 차트 중 4)

# hist : 히스토그램
ax1.hist(np.random.randn(100))

# scatter : 산점도
ax2.scatter(np.arange(30), np.random.randn(30))

# plot : 선 그래프
ax3.plot(np.random.randn(50).cumsum())

# 차트 보기
plt.show()
````

````python
# 쓰지 않는 차트도 같이 보여짐
fig, axes = plt.subplots(2, 2)

axes[0][0].hist(np.random.randn(100))
axes[0][1].scatter(np.arange(30), np.random.randn(30))
axes[1][1].plot(np.random.randn(50).cumsum())

plt.show()
````

#### 차트 꾸미기

````python
fig, axes = plt.subplots(2, 2)

# bins : 범위 (bins=20 : 범위 20개)
# color : 차트 색상
# alpha : 투명도 (0.0[투명] ~ 1.0[불투명])
axes[0][0].hist(np.random.randn(100), bins=20, color='red', alpha=0.5)
axes[0][1].scatter(np.arange(30), np.random.randn(30), color='lightgreen')

# marker : 값 위치 표시
# (1) * (별) / (2) ^ (삼각형) / (3) o (동그라미) / ...
axes[1][0].plot(np.random.randn(50).cumsum(), linestyle=':', color='r', marker='o')

# linestyle : 선 종류 
# (1) -- (점선) / (2) - (실선) / (3) : (점) / (4) -. (선&점)
axes[1][1].plot(np.random.randn(50).cumsum(), linestyle='--')

plt.show()
````

````python
# 범례 : legend
plt.plot(np.random.randn(50).cumsum(), label='x')
plt.plot(np.random.randn(50).cumsum(), label='y')

# loc : 범례 위치
# (1) upper left (좌측 상단) / (2) upper right (우측 상단) 
# (3) best (최적의 위치) / (4) center (중간)
# (5) lower left (좌측 하단) / (6) lower right (우측 하단)
# (7) left (좌측 중간) / (8) right (우측 중간)
plt.legend(loc='right')

plt.show()
````

##### Pandas 데이터프레임, 시리즈

````python
# pandas 시리즈
pd.Series(np.random.randn(1000).cumsum()).plot()
plt.show()
````

````python
# pandas 데이터프레임
# pd.DataFrame(np.random.randn(10, 4).cumsum(0), columns=['A', 'B', 'C', 'D'], index=np.arange(0, 100, 10)).plot()
pd.DataFrame(np.random.randn(10, 4).cumsum(0), columns=['A', 'B', 'C', 'D'], index=np.arange(0, 100, 10)).hist()
plt.show()
````

````python
# 차트 제목 설정시 한글 깨지는 현상
import matplotlib
from matplotlib import font_manager, rc

font_path = 'C:/Windows/Fonts/NanumGothic.ttf'
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)

# 한글 사용시 마이너스(-) 깨지는 현상
matplotlib.rcParams['axes.unicode_minus'] = False 

# --------------------------------------------------------------------------------------

chart = pd.Series(np.random.randn(1000).cumsum()).plot()

# set_xticks, set_yticks : x축, y축 간격 조정
chart.set_xticks([0, 250, 500, 750, 1000])
# chart.set_yticks([0, 25, 50])

# set_title : 차트 제목 (한글은 깨짐)
chart.set_title('랜덤 라인차트')
# chart.set_title('Random LineChart')

# set_xticklabels, set_yticklabels : x축, y축 라벨
# rotation : 글씨 회전
chart.set_xticklabels(['zero', 'one', 'two', 'three', 'four'], rotation=300)

# grid(True) : 차트에 그리드 표시 O
chart.grid(True)

plt.show() # 쓸데없는 타입, 결과 메시지 삭제
````

#### 예제 연습

캐글 타이타닉 연습문제 CSV   
https://raw.githubusercontent.com/developer-sdk/kaggle-python-beginner/master/datas/kaggle-titanic/train.csv

````python
train = pd.read_csv('https://raw.githubusercontent.com/developer-sdk/kaggle-python-beginner/master/datas/kaggle-titanic/train.csv')

train.tail()
````

````python
# 성별 생존률 교차도표 생성
gender_survived = pd.crosstab(train['Sex'], train['Survived'])
gender_survived.columns = gender_survived.columns.map({0:"Dead", 1:"Alive"})
````

````python
# bar : 바 차트
# bar = gender_survived.plot.bar()
# bar.set_title('생존자/사망자 수')
bar = gender_survived.plot(kind='bar', title='생존자/사망자 수')

# 차트의 수치 표시
for p in bar.patches:
    left, bottom, width, height = p.get_bbox().bounds
    bar.annotate(f'{height:n}', (left+width/2, height*1.02), ha='center')

gender_survived.plot.bar(subplots=True) # 성별로 따로 바차트를 추가 생성

# 차트는 사망자, x축에는 생존자 수치 표시 (차트와 x라벨을 교차 사용 => 쓸 일 없음!!)
# gender_survived.plot.bar(x='Alive') 

plt.show()
````

````python
# 가로 바차트
gender_survived.plot(kind='barh')
gender_survived.plot.barh(subplots=True)

plt.show()
````

````python
# 라인 차트
gender_survived.plot(kind='line')

plt.show()
````

````python
# 파이 차트
gender_survived['Alive'].plot(kind='pie', title='생존자 비율', autopct='%1.1f%%')

plt.show()
````

````python
gender_survived['Dead'].plot(kind='pie', title='사망자 비율', autopct='%1.1f%%')

plt.show()
````
