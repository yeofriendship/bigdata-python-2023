##### 전처리

```python
import pandas as pd
import numpy as np

import folium
import json
```

```python
# 스타벅스 매장 리스트
sbuck_df = pd.read_excel('./seoul_starbucks.xlsx', header=0)
sbuck_df
```

```python
sbuck_df['주소'][0].split()[1]
```

```python
gu_list = []

for item in sbuck_df['주소']:
    gu = item.split()[1]
    gu_list.append(gu)

sbuck_df['구'] = gu_list
sbuck_df
```

```python
seoul_gu_df = seoul_loc_df[seoul_loc_df['do'] == '서울']
seoul_gu_df
```

#### 서울시 구별 스타벅스 매장 수 구하기

```python
sbuck_gu_counts = sbuck_df.pivot_table(index='구',
                                       values='매장명',
                                       aggfunc='count').rename(columns={'매장명':'스타벅스_매장수'})
sbuck_gu_counts.head()
```

```python
# seoul_gu_df 수정
seoul_gu_df = seoul_gu_df[['do', 'city', 'latitude', 'longitude']]
columns = {'do':'도', 'city':'구', 'latitude':'위도', 'longitude':'경도'}
seoul_gu_df.rename(columns=columns, inplace=True)
seoul_gu_df.reset_index(drop=True, inplace=True)
seoul_gu_df
```

```python
seoul_sgg = pd.merge(seoul_gu_df, sbuck_gu_counts, how='left', on='구')
seoul_sgg
```

```python
seoul_pop = pd.read_excel('./seoul_pop.xlsx')
seoul_pop
```

```python
seoul_sgg = pd.merge(seoul_sgg, seoul_pop, how='left', on='구')
seoul_sgg
```

```python
seoul_sgg = seoul_sgg[['도', '구', '경도', '위도', '스타벅스_매장수', '계']]
seoul_sgg
```

```python
# 사업체 수
seoul_biz = pd.read_excel('./seoul_biz.xlsx')
seoul_biz
```

```python
seoul_sgg_stat = pd.merge(seoul_sgg, seoul_biz, how='left', on='구')
seoul_sgg_stat
```

```python
columns = {'도':'도', '구':'구', '경도':'경도', '위도':'위도', '스타벅스_매장수':'스타벅스_매장수', 
           '계':'인구수', '사업체수':'사업체수', '종사자수 합계':'종사자수'}
seoul_sgg_stat.rename(columns=columns, inplace=True)
seoul_sgg_stat
```

```python
seoul_sgg_stat.to_excel('./seoul_sgg_stat.xlsx', index=False)
```

#### 데이터 시각화

```python
sbuck_df.tail()
```

```python
sbuck_map = folium.Map(location=[37.550823, 126.989502], tiles='Stamen Terrain', zoom_start=11)

for idx in sbuck_df.index:
    lat = sbuck_df.loc[idx, '위도']
    lng = sbuck_df.loc[idx, '경도']
    name = sbuck_df.loc[idx, '매장명']

    folium.CircleMarker(location=[lat, lng], fill=True, fill_color='green', fill_opacity=1,
                        popup=name, color='yellow', weight=1, radius=3).add_to(sbuck_map)

sbuck_map
```

```python
sbuck_map2 = folium.Map(location=[37.550823, 126.989502], tiles='Stamen Terrain', zoom_start=11)

for idx in sbuck_df.index:
    lat = sbuck_df.loc[idx, '위도']
    lng = sbuck_df.loc[idx, '경도']
    name = sbuck_df.loc[idx, '매장명']
    type = sbuck_df.loc[idx, '매장 타입']

    # 매장 타입별 색상
    fillcolor = ''
    size = 2
    if type == 'general':
        fillcolor = 'gray'
        size = 2
    elif type == 'reserve':
        fillcolor = 'black'
        size = 5
    elif type == 'generalDT':
        fillcolor = 'red'
        size = 3
    elif type == 'generalWT':
        fillcolor = 'blue'
        size = 3

    folium.CircleMarker(location=[lat, lng], fill=True, fill_color=fillcolor, fill_opacity=1,
                        popup=name, color='yellow', weight=1, radius=size).add_to(sbuck_map2)

sbuck_map2
```

```python
sbuck_mean = seoul_sgg_stat['스타벅스_매장수'].mean()
sbuck_mean
```

```python
# 시군구 경계 표시
sgg_geojson_path = './seoul_geo_simple.json'
seoul_sgg_geo = json.load(open(sgg_geojson_path, encoding='utf-8'))

sbuck_map4 = folium.Map(location=[37.550823, 126.989502], tiles='CartoDB dark_matter', zoom_start=11)

folium.GeoJson(seoul_sgg_geo, name='Seoul 행정구역').add_to(sbuck_map4)

for idx in seoul_sgg_stat.index:
    lat = seoul_sgg_stat.loc[idx, '위도']
    lng = seoul_sgg_stat.loc[idx, '경도']
    count = seoul_sgg_stat.loc[idx, '스타벅스_매장수']

    if count > sbuck_mean: # 각 구별 평균치보다 매장수가 많으면
        fillcolor = 'red'
    else:
        fillcolor = 'blue'
    
    folium.CircleMarker(location=[lat, lng], color='#ffff00', fill_color=fillcolor, 
                        fill_opacity=0.7, weight=1.5, radius=count/2, popup=count).add_to(sbuck_map4)

sbuck_map4
```

```python
# 시군구 경계 표시
sgg_geojson_path = './seoul_geo_simple.json'
seoul_sgg_geo = json.load(open(sgg_geojson_path, encoding='utf-8'))

sbuck_map5 = folium.Map(location=[37.550823, 126.989502], tiles='CartoDB dark_matter', zoom_start=11)

folium.GeoJson(seoul_sgg_geo, name='Seoul 행정구역').add_to(sbuck_map5)

folium.Choropleth(
    geo_data=seoul_sgg_geo,
    data=seoul_sgg_stat,
    columns=['구', '스타벅스_매장수'],
    fill_color='Set1', # YlGn / RdYlGn / Spectral / Accent / Reds / Greends / Blues / Oranges / PuOr / Set1
    fill_opacity=0.7,
    line_opacity=0.5,
    key_on='properties.name' # key 값이 없으면 지도에 색상표시 못함!!
).add_to(sbuck_map5)

sbuck_map5
```

```python
# 서울시 인구수별 분포도
seoul_pop_choropleth = folium.Map(location=[37.550823, 126.989502], tiles='CartoDB dark_matter', zoom_start=10)

folium.Choropleth(geo_data=seoul_sgg_geo, data=seoul_sgg_stat,
                  columns=['구', '인구수'], fill_color='Purples',
                  fill_opacity=0.7, line_opacity=0.5,
                  key_on='properties.name').add_to(seoul_pop_choropleth)

seoul_pop_choropleth
```

- 결론 : 구에 살고 있는 인구수와 스타벅스 매장수는 거의 관련이 없다.
