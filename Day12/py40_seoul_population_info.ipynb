```python
import warnings
warnings.simplefilter("ignore")

# 서울시 시군구별 인구 데이터
pop_df = pd.read_excel('./주민등록인구_202307.xlsx', header=1, usecols='B:H')
columns = {'동별(2)':'구', '세대 (세대)':'세대수', '계':'계', '한국인':'한국인', 
           '등록외국인':'외국인', '세대당인구':'비율', '65세이상고령자':'고령자'}
pop_df.rename(columns=columns, inplace=True)
pop_df
```

```python
# 첫번째 행(합계) 삭제
pop_df.drop([0], axis=0, inplace=True)
pop_df
```

```python
# 인덱스 재정의
pop_df.reset_index(drop=True, inplace=True)
pop_df
```

```python
seoul_last_df = pop_df[['구', '계']]
seoul_last_df.info()
```

```python
seoul_last_df.to_excel('./seoul_pop.xlsx', index=False)
```

##### 서울시 시군구별 사업체 데이터 

```python
biz_df = pd.read_excel('./사업체현황_2307.xlsx', header=3, usecols='B:G')
columns = {
    '동별(2)':'구',
    '소계':'사업체수',
    '여성대표자':'여성대표자',
    '소계.1':'종사자수 합계',
    '남':'종사자(남)',
    '여':'종사자(여)'
}
biz_df.rename(columns=columns, inplace=True)
biz_df
```

```python
# 첫번째 행 삭제
biz_df.drop([0], axis=0, inplace=True)
biz_df
```

```python
seoul_biz_df = biz_df[['구', '사업체수', '종사자수 합계']]
seoul_biz_df.info()
```

```python
seoul_biz_df.to_excel('./seoul_biz.xlsx', index=False)
```
